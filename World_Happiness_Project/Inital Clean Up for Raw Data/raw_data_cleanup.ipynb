```python
## Drop first 3 rows without data and resent the index
education_df = education_df.drop([0,1,2,3])
education_df = education_df.reset_index(drop=True)
education_df
```

```python
## create a new column called Rate to hold the average enrollment rate for columns 2 to 9
## Please note that unnamed: is year starting with 2015 and ending 2020

education_df['Rate'] = ''

for i, rows in education_df.iterrows():
    education_df.loc[i, 'Rate'] = rows[4:9].mean()
        
education_df.head()
```

```python
## make new df with country and enrollment rate and rename columns for export to csv
new_edu_df = education_df[['Data Source','Rate']]
new_edu_df = new_edu_df.rename(columns={'Data Source':'Country'})
new_edu_df
```
